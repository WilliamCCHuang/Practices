```python
from keras.preprocessing import sequence
from keras.preprocessing.text import Tokenizer
```

```python
import re
def rm_tag(text):
    re_tag = re.compile(r'[^>]+>')
    return re_tag.sub('',text)
```

```python
import os
def read_files(file_type):
    path = "data/aclImdb/"
    file_list = []
    
    positive_path = path + file_type + "/pos/"
    for f in os.listdir(positive_path):
        file_list += [positive_path + f]
    
    negative_path = path + file_type + "/neg/"
    for f in os.listdir(negative_path):
        file_list += [negative_path + f]
    
    print("read", file_type, "files:", len(file_list))
    
    all_labels = ([1]*12500 + [0]*12500)
    
    all_texts = []
    for fi in file_list:
        with open(fi, encoding="utf8") as file_input:
            all_texts += [rm_tag(" ".join(file_input.readlines()))]
    
    return all_labels, all_texts
```

```python
y_train, train_text = read_files("train")
```

```python
y_test, test_text = read_files("test")
```

```python
token = Tokenizer(num_words=2000)
token.fit_on_texts(train_text)
```

```python
x_train_seq = token.texts_to_sequences(train_text)
x_test_seq = token.texts_to_sequences(test_text)
```

```python
x_train = sequence.pad_sequences(x_train_seq, maxlen=100)
x_test  = sequence.pad_sequences(x_test_seq,  maxlen=100)
```

```python
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation, Flatten
from keras.layers.embeddings import Embedding
```

```python
model = Sequential()
```

```python
model.add(Embedding(output_dim=32, input_dim=2000, input_length=100))
```

```python
model.add(Dropout(0.2))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(units=256, activation="relu"))
```

```python
model.add(Dropout(0.35))
```

```python
model.add(Dense(units=1, activation="sigmoid"))
```

```python
model.summary()
```

```python
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
train_history = model.fit(x_train, y_train, batch_size=100, epochs=10, verbose=2, validation_split=0.2)
```

```python
model.evaluate(x_test, y_test, verbose=1)
```

```python
predict = model.predict_classes(x_test)
```
